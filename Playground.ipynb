```python
import requests
import json
import pandas as pd


# Replace 'url' with the actual URL from which you're fetching the JSON
#url = 'https://bj.scjn.gob.mx/api/buscador/documento/tesis/35uYz3YBN_4klb4Hy8ap'
#"https://bj.scjn.gob.mx/doc/tesis/7jQMbHkBNHmckC8LBcui/*"
url = "https://bj.scjn.gob.mx/api/buscador/documento/tesis/7jQMbHkBNHmckC8LBcui"
#url = "https://bj.scjn.gob.mx/api/buscador/documento/tesis/__6g8YYBAeINReW6jUpQ"

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response into a Python dictionary
    data = response.json()

    # Pretty print the JSON
    formatted_json = json.dumps(data, indent=4, ensure_ascii=False)
    print(formatted_json)
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
def flatten_json(b, parent_key='', delim='_'):
    items = {}
    for key, value in b.items():
        new_key = f"{parent_key}{delim}{key}" if parent_key else key
        if isinstance(value, dict):
            items.update(flatten_json(value, new_key, delim=delim))
        else:  # it's a list of any kind or a single value
            items[new_key] = value
    return items

json_obj = data

# Flatten the JSON object
flat = flatten_json(json_obj)

#print(flat)

# Create a DataFrame with the flattened data
# Since we are creating a DataFrame from a single record, we need to pass [flat]
df = pd.DataFrame([flat])

df.rename(columns={'id': 'idBuscadorJuridico'}, inplace=True)


print(df)
```
